```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path="/content/drive/MyDrive/Kaggle/Titanic/csv"
```

##carga de los datos

```python
gender_submission=pd.read_csv(path+"/gender_submission.csv")
test=pd.read_csv(path+"/test.csv")
train=pd.read_csv(path+"/train.csv")
```

```python
pasId=test["PassengerId"]
```

##eliminacion de columnas inecesarias

```python
for column in ["PassengerId","Name","Ticket","Cabin"]:
  train.pop(column)
  test.pop(column)
```

```python
display(train)
```

```python
display(test)
```

##separacion de X y Y 

```python
trainY = pd.DataFrame(train.pop("Survived")) 
display(trainY)
```

##eliminacion de NAN variables dummy y demas

```python
trainData=train.iloc[:,:].values
testData=test.iloc[:,:].values
```

```python
display(pd.DataFrame(trainData))
```

```python
display(pd.DataFrame(testData))
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = "mean", verbose=0)
imputer = imputer.fit(trainData[:,2:3]) 
trainData[:,2:3] = imputer.transform(trainData[:,2:3])
```

```python
imputer=imputer.fit(testData[:,2:3])
testData[:,2:3] = imputer.transform(testData[:,2:3])
```

```python
imputer=imputer.fit(testData[:,5:6])
testData[:,5:6] = imputer.transform(testData[:,5:6])
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy="most_frequent",verbose=0)
imputer = imputer.fit(trainData[:,6:7]) 
trainData[:,6:7] = imputer.transform(trainData[:,6:7])
```

```python
display(pd.DataFrame(trainData))
```

```python
display(pd.DataFrame(testData))
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
labelencoder = LabelEncoder()
trainData[:, 1] = labelencoder.fit_transform(trainData[:, 1])
testData[:, 1] = labelencoder.fit_transform(testData[:, 1])
```

```python
display(pd.DataFrame(trainData))
```

```python
display(pd.DataFrame(testData))
```

```python
ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(categories='auto'), [6])],   
    remainder='passthrough'                        
)
```

```python
testData = ct.fit_transform(testData)
trainData = ct.fit_transform(trainData)
```

```python
display(pd.DataFrame(trainData))
```

```python
display(pd.DataFrame(testData))
```

##escalado de datos

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
trainData = sc.fit_transform(trainData)
testData = sc.transform(testData)
```

##aplicamos modelo 

```python
trainDataY=trainY.iloc[:,:].values
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(n_estimators = 100, criterion = "entropy", random_state = 0)
classifier.fit(trainData,trainDataY )
```

##prediccion

```python
y_pred  = classifier.predict(testData)
final=pd.DataFrame()
final["PassengerId"]=pasId
final["Survived"]=y_pred
display(final)
```

```python
final.to_csv(path+"/final.csv",index=False)
```
